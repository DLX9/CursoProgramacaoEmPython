```python
from re import escape
#Variaveis do algoritimo
nome = "nome do usuário"
nota1 = 0
nota2 = 0
media = 0

#Recupera o nome do usuário
nome = input("Digite seu nome: ")
print("Bem vindo(a) ao sistema de calcular média " + nome)

#Recupera a nota do usuário
nota1 = float(input("Por favor digite a sua primeira nota: "))
nota2 = float(input("Por favor digite a sua segunda nota: "))

#Calcula média do usuário
media = round((nota1 + nota2)/2, 1)

#Exibe ao usuário a sua média
print("Processando...")
print(f"Sua média do é: {media}")

#Sisteam condicional que verifica se o usuário foi aprovado ou reprovado
if(media <= 7 and media >= 5.5):
    print("E você está de recuperação")
elif(media >= 7):
    print("E você foi aprovado")
else:
    print("E você foi reprovado")
```
