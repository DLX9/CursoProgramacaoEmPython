```python
#Desafio 2
user = "asd@gmail.com"
password = "asd"
t = 0
n = 2
while t < 3:
  usuario = input("Digite seu email: ")
  senha = input("Digite sua senha: ")
  if(user == usuario) and (password == senha):
    print("Seja bem vindo")
    t += 4
  elif(n > 0):
    print(f"Credenciais incorretas, você tem mais {n} tentativas")
    n -= 1
    t += 1
  else:
    print("Você foi bloqueado por 60 minutos")
    t += 4
```

```python
#Desafio 3
print("Exemplo: quero que calcule todos os números pares de (base) até (maximo)")
n1 = int(input("Digite a base da soma: "))
n2 = int(input("Digite qual o número maximo: "))
r = 0

while n1 <= n2:
  if(n1%2 == 0):
    r += n1
    n1 += 1
  else:
    n1 += 1

print(f"A soma de todos os numeros pares entre: {r}")
```

```python
#Desafio 4
print("Exemplo: quero que calcule todos os números primos de (base) até (maximo)")
n1 = int(input("Digite a base da soma: "))
n2 = int(input("Digite qual o número maximo: "))
r = 0

while n1 <= n2:
  if (n1 == 0) or (n1 == 1):
    n1 += 1
  elif (n1%2 == 0 and n1 != 2) or (n1%3 == 0 and n1 != 3) or (n1%5 == 0 and n1 != 5) or (n1%7 == 0 and n1 != 7) or (n1%11 == 0 and n1 != 11):
    n1 += 1
  else:
    r += n1
    n1 += 1


print(f"A soma de todos os numeros primos é: {r}")
```

```python
#Desafio 5

a = 1
b = 1
print("Lembre que a cada número que inserir você será multiplicado por 3")
x = int(input("Insira quantos numeros da sequencia de fibonate você quer ver: "))

while x > 0:
  print(a)
  print(b)
  c = a + b
  print(c)
  a = c + b
  b = a + c
  x -= 1
```
