```python
#Printando listas
#Listas
frutas = ["Morango", "Mexirica", "Laranja", "Abacaxi", "Uva"]

#Acessa o primeiro elemento
print(frutas[0])
print(frutas[-1])
```

```python
#Operações com listas
#Listas
l1 = [1, 2, 3]
l2 = [4, 5, 6]

conc = l1 + l2
rep = l1 * 2
listo = (l1 * 2) + (l2 * 2)
print(conc)
print(rep)
print(listo)
```

```python
#Manipulação de lista
#Listas
frutas = ["Morango", "Mexirica", "Laranja", "Abacaxi", "Uva"]

print(frutas)

#Adiciona valor no final
frutas.append("Amora")

print(frutas)

#Remove um elemento
frutas.pop(-1)

print(frutas)

print(frutas.index("Laranja"))
```

```python
#Definição de lista aninhada / agrupada
#Listas
matriz = [
    ["Morango", "Mexirica", "Laranja", "Abacaxi", "Uva"],
    ["Morangooo", "Mexiricaaa", "Laranjaaa", "Abacaxiii", "Uvaaa"]
  ]

print(matriz[1][3])

#Append em matriz
matriz[0].append("amora")
matriz[1].append("amoraaa")
print(matriz)

#Pop em matriz
matriz[0].pop(3)
matriz[1].pop(2)
print(matriz)

#Remove de acordo com o index do elemento
print("")
matriz[1].pop(matriz[1].index("Uvaaa"))
#Remove de acordo com o nome
matriz[0].remove("Mexirica")

print(matriz)
```
